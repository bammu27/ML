```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.model_selection import train_test_split



```

```python
df = pd.read_csv('california_housing.csv')
```

```python
df.head()
```

```python

df['pred0'] = df['MedHouseVal'].mean()
df['res0'] = df['MedHouseVal'].values - df['pred0']

x = df.iloc[:, :8].values

for i in range(1, 7):
    y = df.iloc[:, -1].values  # residue
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
    clf = DecisionTreeRegressor()
    clf.fit(x_train, y_train)
    
    # Fix the list comprehension syntax
    df[f'pred{i}'] = clf.predict(x) * 0.01 + np.sum([0.01 * df[f'pred{j}'] for j in range(1, i)])
    
    df[f'res{i}'] = df['MedHouseVal'].values - df[f'pred{i}']
```
