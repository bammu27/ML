```python
 release_data= df.groupby(by=['release_year']).mean()
```

```python
release_data
```

WHICH IS BEST YEARs FOR MOVIE

```python
release_data[(release_data['imdb_score']>release_data['imdb_score'].quantile(0.75))&(release_data['imdb_votes']>release_data['imdb_votes'].quantile(0.9))]
```

TOP MOVIES or SHOWS OF EACH YEAR

```python
grouped_data = df.groupby('release_year')
```

```python
def topmovie(x):
    y =  x.sort_values(by='imdb_score',ascending=False)
    return y[y['imdb_votes']>y['imdb_votes'].quantile(0.9)].head(1)


top_movie_byYear = grouped_data.apply(topmovie)
```
