```python
import pandas as pd
df = pd.read_csv('iris.csv')

df
```

```python
#knn from scartch
import numpy as np
import pandas as  pd
from collections import Counter


def eculdian_distance(x,y):
    return np.sqrt(np.sum((x-y)**2))


class KNN:
    def __init__(self,k):
        self.k = k
    
    def fit(self,x_train,y_train):
        self.x_train = x_train
        self.y_train = y_train

    def predict(self,X):
        distance = [self.calculate(x) for x in X]

        indices = np.argsort(distance,axis=1)[:,:self.k]

        species_list = [self.y_train.iloc[index].tolist() for index in indices]
        counts = [Counter(species).most_common() for species in species_list]

        predication = []

        for target in counts:
            predication.append(target[0][0])


        return predication

        

    def calculate(self,x):
        x_list = [eculdian_distance(x,x_t) for x_t in self.x_train]
        return x_list
```

```python
from sklearn.model_selection import train_test_split

X = np.array(df.iloc[:,:-1])
Y = df.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
knn = KNN(5)

knn.fit(x_train,y_train)

prediction = knn.predict(x_test)
actual = np.array(y_test)

accuracy = np.sum(prediction==actual)/len(y_test) *100
```
